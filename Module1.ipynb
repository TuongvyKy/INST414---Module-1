```python
import pandas as pd
import matplotlib.pyplot as plt
```

Maryland Temperature Data

```python
# Annual Average Temperature from 1990 - 2023
annual_df = pd.read_csv('annual_avg_temperature.csv')
annual_df.rename(columns = {'Date':'Year', 'Value': 'Avg_Temperature(F)' }, inplace = True)
annual_df['Year'] = annual_df['Year'] // 100
```

```python
# Line Graph of Annual Average Temperature from 1990 - 2023
plt.plot(annual_df['Year'], annual_df['Avg_Temperature(F)'])
plt.title('Maryland Annual Average Temperature from 1900 to 2023')
plt.xlabel('Year')
plt.ylabel('Average Temperatre(F)')
plt.show
```

```python
# Line Graph of Annual Average Temperature Anomalies from 1990 - 2023 using a Mean of 53.8F
plt.plot(annual_df['Year'], annual_df['Anomaly'])
plt.title('Maryland Annual Average Temperature from 1900 to 2023 Anomalies (Mean:53.8F)')
plt.xlabel('Year')
plt.ylabel('Anomaly')
plt.show
```
